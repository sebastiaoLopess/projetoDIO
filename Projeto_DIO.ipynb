# Extração de Dados

```python
import pandas as pd
import pandera as pa
plt.style.use("seaborn")
```

```python
df = pd.read_csv('URUBURETAMA-RES-RESULTADO_GERAL.csv', sep = ';')
```

```python
df.set_index('Numero', inplace = True)
```

```python
df['VotosNominais'] = df['VotosNominais'].astype('int64')
```

```python
df.dtypes
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "VotosNominais": pa.Column(pa.Int),
    }
)
```

```python
schema.validate(df)
```

# Análise Exploratória

### Qual o total de votos na eleição?

```python
total_votos = df.VotosNominais.sum()
total_votos
```

### Qual o total de votos por partido?

```python
import matplotlib.pyplot as plt
df.groupby(['Partido'])['VotosNominais'].sum().sort_values(ascending=False).plot.bar(title = "Votos por partido");
plt.xlabel("Partido")
plt.ylabel("Votos");
```

### Qual a quantidade de candidatos por partido

```python
import matplotlib.pyplot as plt
df.groupby(['Partido'])['NomeDoCandidato'].count().sort_values(ascending=False).plot.bar(title = "Qtd de candidatos por partido");
plt.xlabel("Partido")
plt.ylabel("Qtd Candidatos");
```

### Qtd de candidatos eleitos x não eleitos

```python
df.groupby(['Situacao'])['NomeDoCandidato'].count().sort_values(ascending=False).plot.pie(title = "Qtd Eleitos e Nao Eleitos");
```

### Qual a quantidade de candidaturas por cargos

```python
df.groupby(['Cargo'])['NomeDoCandidato'].count().sort_values(ascending=False).plot.bar(title = "Cargos");
```
